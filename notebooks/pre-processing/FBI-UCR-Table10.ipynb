```python
import pandas as pd 
import numpy as np 
```

```python
def ParseExcel(Filename,skiprow_val = 3 ,skipfooter_val = 4):
    return pd.read_excel(Filename,skiprows=skiprow_val,skipfooter = skipfooter_val)

Table10_DF_2010 = ParseExcel(Table10_location + "2010.xls",3,2)
Table10_DF_2010["Year"] = 2010
Table10_DF_2010

# Next Year 2011 
Table10_DF_2011 = ParseExcel(Table10_location + "2011.xls",4,3)
Table10_DF_2011["Year"] = 2011 
Table10_DF_2011

# Next Year 2012
Table10_DF_2012 = ParseExcel(Table10_location + "2012.xls",5,3)
Table10_DF_2012["Year"] = 2012 
Table10_DF_2012

# Next Year 2013
Table10_DF_2013 = ParseExcel(Table10_location + "2013.xls",5,3)
Table10_DF_2013["Year"] = 2013 
Table10_DF_2013

# Next Year 2014
Table10_DF_2014 = ParseExcel(Table10_location + "2014.xls",5,3)
Table10_DF_2014["Year"] = 2014 
Table10_DF_2014

# Next Year 2015
Table10_DF_2015 = ParseExcel(Table10_location + "2015.xls",5,3)
Table10_DF_2015["Year"] = 2015 
Table10_DF_2015

# Next Year 2016
Table10_DF_2016 = ParseExcel(Table10_location + "2016.xls",5,3)
Table10_DF_2016["Year"] = 2016 
Table10_DF_2016

# Next Year 2017
Table10_DF_2017 = ParseExcel(Table10_location + "2017.xls",5,3)
Table10_DF_2017["Year"] = 2017
Table10_DF_2017

# Next Year 2018
Table10_DF_2018 = ParseExcel(Table10_location + "2018.xls",5,3)
Table10_DF_2018["Year"] = 2018
Table10_DF_2018

# Next Year 2019
Table10_DF_2019 = ParseExcel(Table10_location + "2019.xls",5,3)
Table10_DF_2019["Year"] = 2019
Table10_DF_2019
```

```python
Table_list = [
                Table10_DF_2010,
                Table10_DF_2011,
                Table10_DF_2012,
                Table10_DF_2013,
                Table10_DF_2014,
                Table10_DF_2015,
                Table10_DF_2016,
                Table10_DF_2017,
                Table10_DF_2018,
                Table10_DF_2019
            ]
```

```python
for table in Table_list:
    table.rename(columns={'Unnamed: 0':"Location", 'Unnamed: 1':"Total incidents"},inplace=True)
    
    if 'Ethnicity/\nnational\norigin' in table.columns:
        table.rename(columns={'Ethnicity/\nnational\norigin':'Ethnicity'},inplace=True)
        
    if 'Race/\nEthnicity/\nAncestry'in table.columns:
        table.rename(columns={'Race/\nEthnicity/\nAncestry':'Ethnicity'},inplace=True)

    if 'Gender\nIdentity' in table.columns:
        table.rename(columns={'Gender\nIdentity':'Gender Identity'},inplace=True)

    print(table.columns)
```

```python
FBI_UCR_TABLE10_FINALIZED = pd.concat(Table_list)
```

```python
FBI_UCR_TABLE10_FINALIZED

FBI_UCR_TABLE10_FINALIZED = FBI_UCR_TABLE10_FINALIZED.loc[:,~FBI_UCR_TABLE10_FINALIZED.columns.str.startswith('Unnamed')]
```
