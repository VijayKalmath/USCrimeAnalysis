```python
import pandas as pd 
import numpy as np 
```

```python
def ParseExcel(Filename,skiprow_val = 3 ,skipfooter_val = 4):
    return pd.read_excel(Filename,skiprows=skiprow_val,skipfooter = skipfooter_val)

Table8_DF_2010 = ParseExcel(Table8_location + "2010.xls",3,3)
Table8_DF_2010["Year"] = 2010
Table8_DF_2010.rename(columns={'Religious\norganization':'Religious \norganization'},inplace=True)

# Next Year 2011 
Table8_DF_2011 = ParseExcel(Table8_location + "2011.xls",4,3)
Table8_DF_2011["Year"] = 2011 
Table8_DF_2011

# Next Year 2012
Table8_DF_2012 = ParseExcel(Table8_location + "2012.xls",5,3)
Table8_DF_2012["Year"] = 2012 
Table8_DF_2012

# Next Year 2013
Table8_DF_2013 = ParseExcel(Table8_location + "2013.xls",5,3)
Table8_DF_2013["Year"] = 2013 
Table8_DF_2013

# Next Year 2014
Table8_DF_2014 = ParseExcel(Table8_location + "2014.xls",5,3)
Table8_DF_2014["Year"] = 2014 
Table8_DF_2014

# Next Year 2015
Table8_DF_2015 = ParseExcel(Table8_location + "2015.xls",5,3)
Table8_DF_2015["Year"] = 2015 
Table8_DF_2015

# Next Year 2016
Table8_DF_2016 = ParseExcel(Table8_location + "2016.xls",5,3)
Table8_DF_2016["Year"] = 2016 
Table8_DF_2016

# Next Year 2017
Table8_DF_2017 = ParseExcel(Table8_location + "2017.xls",5,3)
Table8_DF_2017["Year"] = 2017
Table8_DF_2017

# Next Year 2018
Table8_DF_2018 = ParseExcel(Table8_location + "2018.xls",5,3)
Table8_DF_2018["Year"] = 2018
Table8_DF_2018

# Next Year 2019
Table8_DF_2019 = ParseExcel(Table8_location + "2019.xls",5,3)
Table8_DF_2019["Year"] = 2019
Table8_DF_2019
```

```python
Table_list = [
                Table8_DF_2010,
                Table8_DF_2011,
                Table8_DF_2012,
                Table8_DF_2013,
                Table8_DF_2014,
                Table8_DF_2015,
                Table8_DF_2016,
                Table8_DF_2017,
                Table8_DF_2018,
                Table8_DF_2019
            ]
```

```python
for table in Table_list:
    table.rename(columns={'Unnamed: 0':"Bias motivation", 'Unnamed: 1':"Total incidents"},inplace=True)
#     print(table.columns)
Table8_DF_2010.columns
```

```python
FBI_UCR_TABLE8_FINALIZED = pd.concat(Table_list)
```

```python
FBI_UCR_TABLE8_FINALIZED

FBI_UCR_TABLE8_FINALIZED = FBI_UCR_TABLE8_FINALIZED.loc[:,~FBI_UCR_TABLE8_FINALIZED.columns.str.startswith('Unnamed')]
```
