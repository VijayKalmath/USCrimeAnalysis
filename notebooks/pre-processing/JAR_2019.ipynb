```python
import numpy as np
import pandas as pd 
```

```python
def ParseExcel(Filename,skiprow_val = 3 ,skipfooter_val = 4):
    return pd.read_excel(Filename,skiprows=skiprow_val,skipfooter = skipfooter_val)

Jar_df = ParseExcel("raw_data/Juvenile_Arrest_Records.xls",9,9)
```

```python
JuvenileArrest_df = Jar_df.iloc[:21]
JuvenileArrest_df
```

```python
JuvenileArrest_df.to_csv("processed_data/JuvenileArrestRecords/All_JAR_Data.csv", index=False)
```

```python
# Male Data 
Start_value = 24 
End_value  = 45
Jar_Males_df = Jar_df[Start_value:End_value]
Jar_Males_df["Gender"] = "Male"

Start_value = 48 
End_value  = 69
Jar_Females_df =  Jar_df[Start_value:End_value]
Jar_Females_df["Gender"] = "Female"

Jar_Gender_df = pd.concat([Jar_Males_df,Jar_Females_df])
Jar_Gender_df.to_csv("processed_data/JuvenileArrestRecords/JAR_Databy_Gender.csv", index=False)
Jar_Gender_df.shape
```
